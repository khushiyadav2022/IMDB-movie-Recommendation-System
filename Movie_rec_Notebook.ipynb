```python
from sklearn.feature_extraction.text import CountVectorizer
from datasketch import MinHashLSHForest
```

```python
# Assuming your dataset is stored in a pandas DataFrame called "new_df"
tags = df_tag2['tag'].tolist()
```

```python
# Create an instance of CountVectorizer
vectorizer = CountVectorizer()
```

```python
# Fit and transform the text data
tag_vectors = vectorizer.fit_transform(tags)
```

```python
# Initialize MinHashLSHForest
lsh_forest = MinHashLSHForest(num_perm=128)
```

```python
# Build the index for approximate nearest neighbors
for i, vector in enumerate(tag_vectors):
    minhash = MinHash(num_perm=128)
    for term in vector.indices:
        minhash.update(str(term).encode('utf-8'))
    lsh_forest.add(i, minhash)
```

```python
# Index the forest
lsh_forest.index()
```

```python
# Query for nearest neighbors
query_index = 0  # Change this index to query for different movies
query_vector = tag_vectors[query_index]
minhash_query = MinHash(num_perm=128)
for term in query_vector.indices:
    minhash_query.update(str(term).encode('utf-8'))
```

```python
# Specify the number of results to retrieve
num_results = 10

# Perform the query and retrieve the nearest neighbors
nearest_neighbors = lsh_forest.query(minhash_query, top_k=num_results)
```

```python
# Find approximate nearest neighbors
nearest_neighbors = lsh_forest.query(minhash_query, num_results=10)
```

```python
# Print the indices of the nearest neighbors
print(nearest_neighbors)
```
